# Analyzing the Autism Prevalence

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from datetime import datetime
```

Read csv files to Dataframe

```python
# Read Prevalence Estimates by Race_Ethnicity
ethnicity = pd.read_csv("../data/Prevalence Estimates by Race_Ethnicity.csv")
ethnicity.head()
```

```python
# Read Prevalence Estimates by Gender
gender = pd.read_csv("../data/Prevalence Estimates by Sex.csv")
gender.head()
```

```python
gender_ethnicity = pd.merge(gender, ethnicity, on= ["Year","State"], how = "left")
gender_ethnicity
```

```python
ethnicity_gender_year = gender_ethnicity.groupby("Year")[["Boys", "Girls","White","Black","Hispanic", "Asian or Pacific Islander","American Indian or Alaska Native","Multiracial"]].mean().reset_index()
ethnicity_gender_year
```
