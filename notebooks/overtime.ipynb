# Analyzing the Autism Prevalence

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from datetime import datetime
```

Read csv files to Dataframe

```python
# Read Prevalence Estimates Over Time
overtime = pd.read_csv("../data/Prevalence Estimates Over Time.csv")
overtime.head()
```

```python
# Create a line plot to show the trend over the years
plt.figure(figsize=(10, 6), facecolor="black")

# Custom colors
color = ['#12239E','#40DBE6', '#DE6A73', '#E1E194']
sns.lineplot(data=overtime, x="Year", y="Prevalence Estimates per 1,000 Children",palette = color,  hue="Data Set",  linewidth=2.5,
    marker="o",
    markersize=7, legend=True
)

# Make axes, ticks, and spines white
ax = plt.gca()
ax.set_facecolor("black")
#ax.xaxis.label.set_color("white")
#ax.yaxis.label.set_color("white")

# Tick labels (x and y)
ax.tick_params(axis='x', colors='white', labelsize=10)
ax.tick_params(axis='y', colors='white', labelsize=10)

# Remove axis labels
ax.set_xlabel("")
ax.set_ylabel("")

# Force-set tick label text objects (sometimes needed in dark themes)
# for tick in ax.get_xticklabels():
#     tick.set_color("white")
#     tick.set_fontweight("bold")

# for tick in ax.get_yticklabels():
#     tick.set_color("white")
#     tick.set_fontweight("bold")

# Spines
for spine in ax.spines.values():
     spine.set_color("white")
    
#plt.title("Prevalence Rates per 1000 Children by Data Sources",color ='white', size=12, fontweight = 'bold')

# Legend styling
legend = plt.legend(facecolor="navy", edgecolor="white", labelcolor="white")
for text in legend.get_texts():
    text.set_color("white")

plt.tight_layout()
plt.savefig("cases_dataset", dpi=300, bbox_inches="tight", transparent=True)
plt.show()
```

```python
# Read Prevalence Estimates Over Time
overtime_state = overtime.groupby("State")["Prevalence Estimates per 1,000 Children"].mean().reset_index().sort_values(by="Prevalence Estimates per 1,000 Children",ascending=False)
overtime_state.head(10)
```
